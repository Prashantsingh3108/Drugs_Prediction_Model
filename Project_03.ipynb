```python
from sklearn.preprocessing import LabelEncoder
l=['Sex','BP','Cholesterol']
le=LabelEncoder()
```

```python
for i in l:
    df[i]=le.fit_transform(df[i])
```

```python
x=df.drop(columns=['Drug'])
y=df[['Drug']]
```

```python
x=pd.get_dummies(x)
df.shape
x
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x, y, test_size=0.33, random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
se=StandardScaler()
```

```python
x_train=se.fit_transform(x_train)
x_test=se.transform(x_test) 
```

```python
from sklearn.linear_model import LogisticRegression
Logistic_model=LogisticRegression()
Logistic_model.fit(x_train,y_train)
```

```python
y_pred=Logistic_model.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
def evaluate_model(model, x_train, x_test, y_train, y_test):
    train_pred = model.predict(x_train)
    test_pred = model.predict(x_test)

    print(f"Model: {type(model).__name__}")
    print(f"Train Accuracy: {accuracy_score(y_train, train_pred)}")
    print(f"Test Accuracy: {accuracy_score(y_test, test_pred)}")
    print(f"Train Precision: {precision_score(y_train, train_pred, average='weighted')}")
    print(f"Test Precision: {precision_score(y_test, test_pred, average='weighted')}")
    print(f"Train Recall: {recall_score(y_train, train_pred, average='weighted')}")
    print(f"Test Recall: {recall_score(y_test, test_pred, average='weighted')}")
    print(f"Train F1 Score: {f1_score(y_train, train_pred, average='weighted')}")
    print(f"Test F1 Score: {f1_score(y_test, test_pred, average='weighted')}")
```

```python
evaluate_model(Logistic_model, x_train, x_test, y_train, y_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier

knn_classifier = KNeighborsClassifier(n_neighbors=5)

knn_classifier.fit(x_train, y_train)
```

```python
evaluate_model(knn_classifier,x_train, x_test, y_train, y_test)
```

```python
from sklearn.ensemble import GradientBoostingClassifier
grad=GradientBoostingClassifier(n_estimators=100, learning_rate=1.0,
    max_depth=1, random_state=0)
grad.fit(x_train,y_train)
```

```python
evaluate_model(grad,x_train, x_test, y_train, y_test)
```

```python
import joblib
bundle={
     'best_model':grad,
        'scalar':se,
    'feature_columns':x.columns.tolist()
}
```
